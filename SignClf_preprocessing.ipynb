```python

src1 = r'/content/drive/MyDrive/DL_mydataset/img_sign'
des1 = r'/content/drive/MyDrive/DL_mydataset/bw_sign'

file1 = os.listdir(src1)

for pic in file1:
  img = cv2.imread(os.path.join(src1,pic))
  gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
  crp = gray[:,60:560]
  res = cv2.resize(crp,(256,256))
  cv2.imwrite(os.path.join(des1,pic),res)
```

```python

src2 = r'/content/drive/MyDrive/DL_mydataset/img_nosign'
des2 = r'/content/drive/MyDrive/DL_mydataset/bw_nosign'

file2 = os.listdir(src2)

for pic in file2:
  img = cv2.imread(os.path.join(src2,pic))
  gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
  crp = gray[:,40:580]
  res = cv2.resize(crp,(256,256))
  cv2.imwrite(os.path.join(des2,pic),res)
```

```python
dir1 = "/content/drive/MyDrive/DL_mydataset/bw_sign"
dir2 = "/content/drive/MyDrive/DL_mydataset/bw_nosign"

dest1 = "/content/drive/MyDrive/DL_mydataset/diff_sign"
dest2 = "/content/drive/MyDrive/DL_mydataset/diff_nosign"

dir1_list = os.listdir(dir1)
dir2_list = os.listdir(dir2)

def abs_diff(img1, img2):
  absdiff = cv2.absdiff(img1,img2)
  return absdiff
```

```python
for i in range(0,len(dir1_list)-1):
  crnt_img = dir1_list[i]
  next_img = dir1_list[i+1]
  img1 = cv2.imread(os.path.join(dir1,crnt_img))
  img2 = cv2.imread(os.path.join(dir1,next_img))
  absdiff = abs_diff(img1,img2)
  cv2.imwrite(os.path.join(dest1,crnt_img),absdiff)
```

for i in range(0,len(dir2_list)-1):
